```python
import torch
import torch.nn as nn
import torch.nn.functional as F
import torch.optim as optim
from torch.utils.data import Dataset, DataLoader
import torchvision.transforms as transforms
from sklearn.model_selection import train_test_split
import cv2
import numpy as np
import os
import glob
from PIL import Image
import matplotlib.pyplot as plt
from tqdm import tqdm
import albumentations as A
from albumentations.pytorch import ToTensorV2

# Custom Dataset Class
class RetinalPigmentDataset(Dataset):
    def __init__(self, image_paths, mask_paths, transform=None, ground_truth='G1'):
        self.image_paths = image_paths
        self.mask_paths = mask_paths
        self.transform = transform
        self.ground_truth = ground_truth

    def __len__(self):
        return len(self.image_paths)

    def __getitem__(self, idx):
        # Load image
        image = cv2.imread(self.image_paths[idx])
        image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

        # Load mask based on ground truth selection
        if self.ground_truth == 'G2':
            mask_g2_path = self.mask_paths[idx].replace('/G1/', '/G2/')
            if os.path.exists(mask_g2_path):
                mask = cv2.imread(mask_g2_path, cv2.IMREAD_GRAYSCALE)
            else:
                mask = cv2.imread(self.mask_paths[idx], cv2.IMREAD_GRAYSCALE)
        else:  # G1 or default
            mask = cv2.imread(self.mask_paths[idx], cv2.IMREAD_GRAYSCALE)

        # Ensure both image and mask have the same dimensions
        if image.shape[:2] != mask.shape[:2]:
            # Resize mask to match image dimensions
            mask = cv2.resize(mask, (image.shape[1], image.shape[0]), interpolation=cv2.INTER_NEAREST)

        # Normalize mask to 0 and 1
        mask = (mask > 127).astype(np.uint8)

        if self.transform:
            try:
                augmented = self.transform(image=image, mask=mask)
                image = augmented['image']
                mask = augmented['mask']
            except Exception as e:
                # If augmentation fails, fall back to basic transforms
                print(f"Augmentation failed for {self.image_paths[idx]}: {e}")
                # Simple resize and normalize without augmentation
                image = cv2.resize(image, (400, 300))
                mask = cv2.resize(mask, (400, 300), interpolation=cv2.INTER_NEAREST)

                # Convert to tensor
                image = torch.from_numpy(image.transpose(2, 0, 1)).float() / 255.0
                mask = torch.from_numpy(mask).long()

                # Normalize image
                mean = torch.tensor([0.485, 0.456, 0.406]).view(3, 1, 1)
                std = torch.tensor([0.229, 0.224, 0.225]).view(3, 1, 1)
                image = (image - mean) / std

        return image.float(), mask.long()

# Simple Conv Block
class ConvBlock(nn.Module):
    def __init__(self, in_channels, out_channels, kernel_size=3, padding=1):
        super(ConvBlock, self).__init__()
        self.conv = nn.Conv2d(in_channels, out_channels, kernel_size, padding=padding, bias=False)
        self.bn = nn.BatchNorm2d(out_channels)
        self.relu = nn.ReLU(inplace=True)

    def forward(self, x):
        return self.relu(self.bn(self.conv(x)))

# Dense Block - Ultra Simple Version
class UltraSimpleDenseBlock(nn.Module):
    def __init__(self, in_channels, growth_rate=32):
        super(UltraSimpleDenseBlock, self).__init__()

        # First conv
        self.conv1 = ConvBlock(in_channels, growth_rate)

        # Second conv
        self.conv2 = ConvBlock(in_channels + growth_rate, growth_rate)

        # Transition layer to reduce channels
        self.transition = ConvBlock(in_channels + 2 * growth_rate, growth_rate)

    def forward(self, x):
        # First conv
        out1 = self.conv1(x)

        # Concatenate and second conv
        concat1 = torch.cat([x, out1], dim=1)
        out2 = self.conv2(concat1)

        # Final concatenation and transition
        concat2 = torch.cat([x, out1, out2], dim=1)
        final_out = self.transition(concat2)

        return out1, final_out  # Return skip connection and main output

# Ultra-Stable RPS-Net
class UltraStableRPSNet(nn.Module):
    def __init__(self, in_channels=3, num_classes=2, base_channels=32):
        super(UltraStableRPSNet, self).__init__()

        # Initial convolution
        self.initial = ConvBlock(in_channels, base_channels)

        # Encoder
        self.enc1 = UltraSimpleDenseBlock(base_channels, base_channels)
        self.pool1 = nn.MaxPool2d(2)

        self.enc2 = UltraSimpleDenseBlock(base_channels, base_channels)
        self.pool2 = nn.MaxPool2d(2)

        self.enc3 = UltraSimpleDenseBlock(base_channels, base_channels)
        self.pool3 = nn.MaxPool2d(2)

        self.enc4 = UltraSimpleDenseBlock(base_channels, base_channels)
        self.pool4 = nn.MaxPool2d(2)

        # Bottleneck
        self.bottleneck = UltraSimpleDenseBlock(base_channels, base_channels)

        # Decoder
        self.up4 = nn.ConvTranspose2d(base_channels, base_channels, 2, stride=2)
        self.dec4 = UltraSimpleDenseBlock(base_channels * 2, base_channels)  # skip connection

        self.up3 = nn.ConvTranspose2d(base_channels, base_channels, 2, stride=2)
        self.dec3 = UltraSimpleDenseBlock(base_channels * 2, base_channels)

        self.up2 = nn.ConvTranspose2d(base_channels, base_channels, 2, stride=2)
        self.dec2 = UltraSimpleDenseBlock(base_channels * 2, base_channels)

        self.up1 = nn.ConvTranspose2d(base_channels, base_channels, 2, stride=2)
        self.dec1 = UltraSimpleDenseBlock(base_channels * 2, base_channels)

        # Final classification
        self.final = nn.Conv2d(base_channels, num_classes, 1)

    def forward(self, x):
        # Initial
        x0 = self.initial(x)

        # Encoder
        skip1, x1 = self.enc1(x0)
        x1 = self.pool1(x1)

        skip2, x2 = self.enc2(x1)
        x2 = self.pool2(x2)

        skip3, x3 = self.enc3(x2)
        x3 = self.pool3(x3)

        skip4, x4 = self.enc4(x3)
        x4 = self.pool4(x4)

        # Bottleneck
        _, x = self.bottleneck(x4)

        # Decoder
        x = self.up4(x)
        if x.shape[2:] != skip4.shape[2:]:
            x = F.interpolate(x, size=skip4.shape[2:], mode='bilinear', align_corners=False)
        x = torch.cat([x, skip4], dim=1)
        _, x = self.dec4(x)

        x = self.up3(x)
        if x.shape[2:] != skip3.shape[2:]:
            x = F.interpolate(x, size=skip3.shape[2:], mode='bilinear', align_corners=False)
        x = torch.cat([x, skip3], dim=1)
        _, x = self.dec3(x)

        x = self.up2(x)
        if x.shape[2:] != skip2.shape[2:]:
            x = F.interpolate(x, size=skip2.shape[2:], mode='bilinear', align_corners=False)
        x = torch.cat([x, skip2], dim=1)
        _, x = self.dec2(x)

        x = self.up1(x)
        if x.shape[2:] != skip1.shape[2:]:
            x = F.interpolate(x, size=skip1.shape[2:], mode='bilinear', align_corners=False)
        x = torch.cat([x, skip1], dim=1)
        _, x = self.dec1(x)

        # Ensure final output size
        if x.shape[2:] != (300, 400):
            x = F.interpolate(x, size=(300, 400), mode='bilinear', align_corners=False)

        # Final classification
        output = self.final(x)
        return output

# Data Loading Functions
def load_images_and_masks(base_path='/content/RIPS',
                         subjects=['subject1', 'subject2', 'subject3', 'subject4']):
    image_base_path = os.path.join(base_path, 'Original')

    images = []
    masks = []

    for subject in subjects:
        sessions = glob.glob(os.path.join(image_base_path, subject, 'session*'))
        for session in sessions:
            for side in ['left', 'right']:
                img_path = os.path.join(session, side)
                mask_path = img_path.replace('Original', 'G1')

                img_files = sorted(glob.glob(os.path.join(img_path, '*.jpg')))
                mask_files = sorted(glob.glob(os.path.join(mask_path, '*.png')))

                for img_file, mask_file in zip(img_files, mask_files):
                    if os.path.exists(img_file) and os.path.exists(mask_file):
                        images.append(img_file)
                        masks.append(mask_file)

    return images, masks

def get_transforms():
    # Simplified transforms with shape checking disabled
    train_transform = A.Compose([
        A.Resize(300, 400),
        A.HorizontalFlip(p=0.3),  # Reduced probability
        A.VerticalFlip(p=0.3),    # Reduced probability
        A.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225]),
        ToTensorV2(),
    ], is_check_shapes=False)  # Disable shape checking

    val_transform = A.Compose([
        A.Resize(300, 400),
        A.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225]),
        ToTensorV2(),
    ], is_check_shapes=False)  # Disable shape checking

    return train_transform, val_transform

def calculate_class_weights(mask_paths):
    """Calculate class weights for handling class imbalance"""
    background_pixels = 0
    pigment_pixels = 0

    print("Calculating class weights...")
    for mask_path in mask_paths[:min(20, len(mask_paths))]:
        try:
            mask = cv2.imread(mask_path, cv2.IMREAD_GRAYSCALE)
            if mask is not None:
                mask = cv2.resize(mask, (400, 300))
                mask = (mask > 127).astype(np.uint8)

                background_pixels += np.sum(mask == 0)
                pigment_pixels += np.sum(mask == 1)
        except:
            continue

    total_pixels = background_pixels + pigment_pixels

    if pigment_pixels == 0:
        return torch.FloatTensor([1.0, 50.0])  # Heavy weight for pigment class

    # Calculate weights
    weight_background = 1.0
    weight_pigment = background_pixels / pigment_pixels if pigment_pixels > 0 else 50.0

    # Cap the pigment weight to avoid extreme values
    weight_pigment = min(weight_pigment, 100.0)

    print(f"Class weights - Background: {weight_background:.2f}, Pigment: {weight_pigment:.2f}")
    return torch.FloatTensor([weight_background, weight_pigment])

def dice_loss(pred, target, smooth=1.0):
    """Dice loss for segmentation"""
    pred = torch.softmax(pred, dim=1)
    pred = pred[:, 1, :, :]
    target = target.float()

    intersection = (pred * target).sum(dim=(1, 2))
    union = pred.sum(dim=(1, 2)) + target.sum(dim=(1, 2))

    dice = (2.0 * intersection + smooth) / (union + smooth)
    return 1 - dice.mean()

def combined_loss(pred, target, weight=None):
    """Combined Cross-Entropy and Dice Loss"""
    ce_loss = F.cross_entropy(pred, target, weight=weight)
    dice_loss_val = dice_loss(pred, target)
    return ce_loss + 0.5 * dice_loss_val

def calculate_metrics(pred, target):
    """Calculate segmentation metrics"""
    pred = torch.softmax(pred, dim=1)
    pred = (pred[:, 1, :, :] > 0.5).float()
    target = target.float()

    pred_flat = pred.view(-1)
    target_flat = target.view(-1)

    tp = (pred_flat * target_flat).sum()
    fp = (pred_flat * (1 - target_flat)).sum()
    fn = ((1 - pred_flat) * target_flat).sum()
    tn = ((1 - pred_flat) * (1 - target_flat)).sum()

    sensitivity = tp / (tp + fn + 1e-8)
    specificity = tn / (tn + fp + 1e-8)
    precision = tp / (tp + fp + 1e-8)
    accuracy = (tp + tn) / (tp + tn + fp + fn + 1e-8)
    f_score = 2 * tp / (2 * tp + fp + fn + 1e-8)
    dice_score = f_score
    iou_score = tp / (tp + fp + fn + 1e-8)

    return {
        'sensitivity': sensitivity.item(),
        'specificity': specificity.item(),
        'precision': precision.item(),
        'accuracy': accuracy.item(),
        'f_score': f_score.item(),
        'dice_score': dice_score.item(),
        'iou_score': iou_score.item()
    }

def train_model(model, train_loader, val_loader, num_epochs=100, device='cuda'):
    """Training loop"""

    # Calculate class weights
    mask_paths = [train_loader.dataset.mask_paths[i] for i in range(len(train_loader.dataset))]
    class_weights = calculate_class_weights(mask_paths)
    class_weights = class_weights.to(device)

    optimizer = optim.Adam(model.parameters(), lr=0.001, weight_decay=1e-4)
    scheduler = optim.lr_scheduler.ReduceLROnPlateau(optimizer, mode='max', patience=3, factor=0.5)

    best_val_dice = 0.0
    train_losses = []
    val_losses = []
    val_metrics_history = []

    for epoch in range(num_epochs):
        # Training phase
        model.train()
        train_loss = 0.0

        pbar = tqdm(train_loader, desc=f'Epoch {epoch+1}/{num_epochs} [Train]')
        for images, masks in pbar:
            images, masks = images.to(device), masks.to(device)

            optimizer.zero_grad()
            outputs = model(images)
            loss = combined_loss(outputs, masks, weight=class_weights)
            loss.backward()
            optimizer.step()

            train_loss += loss.item()

            # Calculate metrics for display
            with torch.no_grad():
                metrics = calculate_metrics(outputs, masks)
                pbar.set_postfix({
                    'Loss': f'{loss.item():.4f}',
                    'Dice': f'{metrics["dice_score"]:.4f}'
                })

        train_loss /= len(train_loader)

        # Validation phase
        model.eval()
        val_loss = 0.0
        val_metrics = {'sensitivity': 0, 'specificity': 0, 'precision': 0, 'accuracy': 0,
                      'f_score': 0, 'dice_score': 0, 'iou_score': 0}

        with torch.no_grad():
            pbar = tqdm(val_loader, desc=f'Epoch {epoch+1}/{num_epochs} [Val]')
            for images, masks in pbar:
                images, masks = images.to(device), masks.to(device)

                outputs = model(images)
                loss = combined_loss(outputs, masks, weight=class_weights)
                val_loss += loss.item()

                metrics = calculate_metrics(outputs, masks)
                for key in val_metrics:
                    val_metrics[key] += metrics[key]

                pbar.set_postfix({
                    'Loss': f'{loss.item():.4f}',
                    'Dice': f'{metrics["dice_score"]:.4f}'
                })

        val_loss /= len(val_loader)
        for key in val_metrics:
            val_metrics[key] /= len(val_loader)

        train_losses.append(train_loss)
        val_losses.append(val_loss)
        val_metrics_history.append(val_metrics.copy())

        # Print results
        print(f'\nEpoch {epoch+1}/{num_epochs}:')
        print(f'Train Loss: {train_loss:.4f}, Val Loss: {val_loss:.4f}')
        print(f'Val Metrics - Dice: {val_metrics["dice_score"]:.4f}, IoU: {val_metrics["iou_score"]:.4f}')
        print(f'Val Metrics - Sens: {val_metrics["sensitivity"]:.4f}, Spec: {val_metrics["specificity"]:.4f}')

        # Save best model
        if val_metrics['dice_score'] > best_val_dice:
            best_val_dice = val_metrics['dice_score']
            torch.save({
                'epoch': epoch,
                'model_state_dict': model.state_dict(),
                'optimizer_state_dict': optimizer.state_dict(),
                'val_metrics': val_metrics,
                'best_dice': best_val_dice
            }, 'best_ultra_stable_rps_net.pth')
            print(f'🎯 Best model saved with Dice: {best_val_dice:.4f}')

        scheduler.step(val_metrics['dice_score'])
        print('-' * 50)

    return {
        'train_losses': train_losses,
        'val_losses': val_losses,
        'val_metrics_history': val_metrics_history
    }

def safe_model_creation_and_training():
    """Safely create and train the model with extensive error handling"""

    # Set device
    device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
    print(f'Using device: {device}')

    try:
        # Load dataset
        print("Loading dataset...")
        image_paths, mask_paths = load_images_and_masks()
        print(f"Total images: {len(image_paths)}")

        # Split dataset
        train_images, val_images, train_masks, val_masks = train_test_split(
            image_paths, mask_paths, test_size=0.25, random_state=42
        )

        print(f"Training images: {len(train_images)}, Validation images: {len(val_images)}")

        # Get transforms
        train_transform, val_transform = get_transforms()

        # Create datasets
        train_dataset = RetinalPigmentDataset(train_images, train_masks, train_transform, ground_truth='G1')
        val_dataset = RetinalPigmentDataset(val_images, val_masks, val_transform, ground_truth='G1')

        # Create data loaders with error handling
        train_loader = DataLoader(train_dataset, batch_size=2, shuffle=True, num_workers=0)  # Set num_workers=0
        val_loader = DataLoader(val_dataset, batch_size=2, shuffle=False, num_workers=0)    # Set num_workers=0

        # Test data loading
        print("Testing data loading...")
        for images, masks in train_loader:
            print(f"Batch - Images: {images.shape}, Masks: {masks.shape}")
            break

        # Create model
        print("Creating model...")
        model = UltraStableRPSNet(in_channels=3, num_classes=2, base_channels=32)

        # Test model on CPU
        print("Testing model on CPU...")
        model.eval()
        with torch.no_grad():
            test_input = torch.randn(1, 3, 300, 400)
            test_output = model(test_input)
            print(f"Model test - Input: {test_input.shape}, Output: {test_output.shape}")

        # Model parameters
        total_params = sum(p.numel() for p in model.parameters())
        print(f'Model Parameters: {total_params:,}')

        # Move to GPU with error handling
        print("Moving model to GPU...")
        if device.type == 'cuda':
            try:
                model = model.to(device)
                print("✅ Model successfully moved to GPU!")

                # Test GPU forward pass
                with torch.no_grad():
                    test_input_gpu = torch.randn(1, 3, 300, 400).to(device)
                    test_output_gpu = model(test_input_gpu)
                    print(f"✅ GPU forward pass successful: {test_output_gpu.shape}")

            except Exception as e:
                print(f"❌ GPU transfer failed: {e}")
                print("Using CPU instead...")
                device = torch.device('cpu')
                model = model.to(device)

        # Train model
        print(f"\nStarting training on {device}...")
        training_results = train_model(model, train_loader, val_loader, num_epochs=100, device=device)

        # Plot results
        val_dice_scores = [m['dice_score'] for m in training_results['val_metrics_history']]
        val_iou_scores = [m['iou_score'] for m in training_results['val_metrics_history']]

        plt.figure(figsize=(12, 4))

        plt.subplot(1, 3, 1)
        plt.plot(training_results['train_losses'], label='Train Loss')
        plt.plot(training_results['val_losses'], label='Val Loss')
        plt.title('Loss Curves')
        plt.xlabel('Epoch')
        plt.ylabel('Loss')
        plt.legend()
        plt.grid(True)

        plt.subplot(1, 3, 2)
        plt.plot(val_dice_scores, label='Dice Score', marker='o')
        plt.plot(val_iou_scores, label='IoU Score', marker='s')
        plt.title('Validation Metrics')
        plt.xlabel('Epoch')
        plt.ylabel('Score')
        plt.legend()
        plt.grid(True)

        plt.subplot(1, 3, 3)
        final_metrics = training_results['val_metrics_history'][-1]
        metrics = ['Dice', 'IoU', 'F-Score', 'Sensitivity']
        values = [final_metrics['dice_score'], final_metrics['iou_score'],
                  final_metrics['f_score'], final_metrics['sensitivity']]

        plt.bar(metrics, values, color=['skyblue', 'lightgreen', 'orange', 'pink'])
        plt.title('Final Metrics')
        plt.ylabel('Score')
        plt.ylim(0, 1)

        plt.tight_layout()
        plt.savefig('ultra_stable_rps_net_results.png', dpi=300, bbox_inches='tight')
        plt.show()

        # Print summary
        print(f"\n🎯 Training Summary:")
        print(f"Best Dice Score: {max(val_dice_scores):.4f}")
        print(f"Best IoU Score: {max(val_iou_scores):.4f}")
        print(f"Device used: {device}")
        print("Training completed successfully!")

    except Exception as e:
        print(f"❌ Error during training: {e}")
        import traceback
        traceback.print_exc()

if __name__ == "__main__":
    safe_model_creation_and_training()
```

```python
import torch
import torch.nn as nn
import torch.nn.functional as F
import cv2
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
import os
import albumentations as A
from albumentations.pytorch import ToTensorV2

# Copy the model architecture (same as training)
class ConvBlock(nn.Module):
    def __init__(self, in_channels, out_channels, kernel_size=3, padding=1):
        super(ConvBlock, self).__init__()
        self.conv = nn.Conv2d(in_channels, out_channels, kernel_size, padding=padding, bias=False)
        self.bn = nn.BatchNorm2d(out_channels)
        self.relu = nn.ReLU(inplace=True)

    def forward(self, x):
        return self.relu(self.bn(self.conv(x)))

class UltraSimpleDenseBlock(nn.Module):
    def __init__(self, in_channels, growth_rate=32):
        super(UltraSimpleDenseBlock, self).__init__()

        self.conv1 = ConvBlock(in_channels, growth_rate)
        self.conv2 = ConvBlock(in_channels + growth_rate, growth_rate)
        self.transition = ConvBlock(in_channels + 2 * growth_rate, growth_rate)

    def forward(self, x):
        out1 = self.conv1(x)
        concat1 = torch.cat([x, out1], dim=1)
        out2 = self.conv2(concat1)
        concat2 = torch.cat([x, out1, out2], dim=1)
        final_out = self.transition(concat2)
        return out1, final_out

class UltraStableRPSNet(nn.Module):
    def __init__(self, in_channels=3, num_classes=2, base_channels=32):
        super(UltraStableRPSNet, self).__init__()

        self.initial = ConvBlock(in_channels, base_channels)

        self.enc1 = UltraSimpleDenseBlock(base_channels, base_channels)
        self.pool1 = nn.MaxPool2d(2)

        self.enc2 = UltraSimpleDenseBlock(base_channels, base_channels)
        self.pool2 = nn.MaxPool2d(2)

        self.enc3 = UltraSimpleDenseBlock(base_channels, base_channels)
        self.pool3 = nn.MaxPool2d(2)

        self.enc4 = UltraSimpleDenseBlock(base_channels, base_channels)
        self.pool4 = nn.MaxPool2d(2)

        self.bottleneck = UltraSimpleDenseBlock(base_channels, base_channels)

        self.up4 = nn.ConvTranspose2d(base_channels, base_channels, 2, stride=2)
        self.dec4 = UltraSimpleDenseBlock(base_channels * 2, base_channels)

        self.up3 = nn.ConvTranspose2d(base_channels, base_channels, 2, stride=2)
        self.dec3 = UltraSimpleDenseBlock(base_channels * 2, base_channels)

        self.up2 = nn.ConvTranspose2d(base_channels, base_channels, 2, stride=2)
        self.dec2 = UltraSimpleDenseBlock(base_channels * 2, base_channels)

        self.up1 = nn.ConvTranspose2d(base_channels, base_channels, 2, stride=2)
        self.dec1 = UltraSimpleDenseBlock(base_channels * 2, base_channels)

        self.final = nn.Conv2d(base_channels, num_classes, 1)

    def forward(self, x):
        x0 = self.initial(x)

        skip1, x1 = self.enc1(x0)
        x1 = self.pool1(x1)

        skip2, x2 = self.enc2(x1)
        x2 = self.pool2(x2)

        skip3, x3 = self.enc3(x2)
        x3 = self.pool3(x3)

        skip4, x4 = self.enc4(x3)
        x4 = self.pool4(x4)

        _, x = self.bottleneck(x4)

        x = self.up4(x)
        if x.shape[2:] != skip4.shape[2:]:
            x = F.interpolate(x, size=skip4.shape[2:], mode='bilinear', align_corners=False)
        x = torch.cat([x, skip4], dim=1)
        _, x = self.dec4(x)

        x = self.up3(x)
        if x.shape[2:] != skip3.shape[2:]:
            x = F.interpolate(x, size=skip3.shape[2:], mode='bilinear', align_corners=False)
        x = torch.cat([x, skip3], dim=1)
        _, x = self.dec3(x)

        x = self.up2(x)
        if x.shape[2:] != skip2.shape[2:]:
            x = F.interpolate(x, size=skip2.shape[2:], mode='bilinear', align_corners=False)
        x = torch.cat([x, skip2], dim=1)
        _, x = self.dec2(x)

        x = self.up1(x)
        if x.shape[2:] != skip1.shape[2:]:
            x = F.interpolate(x, size=skip1.shape[2:], mode='bilinear', align_corners=False)
        x = torch.cat([x, skip1], dim=1)
        _, x = self.dec1(x)

        if x.shape[2:] != (300, 400):
            x = F.interpolate(x, size=(300, 400), mode='bilinear', align_corners=False)

        output = self.final(x)
        return output

def load_trained_model(checkpoint_path='best_ultra_stable_rps_net.pth', device='cuda'):
    """Load the trained model from checkpoint"""

    # Create model
    model = UltraStableRPSNet(in_channels=3, num_classes=2, base_channels=32)

    # Load checkpoint
    if os.path.exists(checkpoint_path):
        print(f"Loading model from {checkpoint_path}")
        checkpoint = torch.load(checkpoint_path, map_location=device)
        model.load_state_dict(checkpoint['model_state_dict'])

        # Print model info
        print(f"✅ Model loaded successfully!")
        print(f"Epoch: {checkpoint['epoch']}")
        print(f"Best Dice Score: {checkpoint['best_dice']:.4f}")
        if 'val_metrics' in checkpoint:
            metrics = checkpoint['val_metrics']
            print(f"Validation Metrics:")
            print(f"  - Dice: {metrics['dice_score']:.4f}")
            print(f"  - IoU: {metrics['iou_score']:.4f}")
            print(f"  - Sensitivity: {metrics['sensitivity']:.4f}")
            print(f"  - Specificity: {metrics['specificity']:.4f}")
    else:
        print(f"❌ Checkpoint not found: {checkpoint_path}")
        print("Please ensure you have trained the model first.")
        return None

    model = model.to(device)
    model.eval()
    return model

def preprocess_image(image_path, target_size=(300, 400)):
    """Preprocess a single image for inference"""

    # Load image
    image = cv2.imread(image_path)
    if image is None:
        raise ValueError(f"Could not load image: {image_path}")

    # Convert BGR to RGB
    image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
    original_image = image.copy()

    # Create transform (same as validation transform)
    transform = A.Compose([
        A.Resize(target_size[0], target_size[1]),
        A.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225]),
        ToTensorV2(),
    ])

    # Apply transform
    transformed = transform(image=image)
    image_tensor = transformed['image'].unsqueeze(0)  # Add batch dimension

    return image_tensor, original_image

def predict_single_image(model, image_path, device='cuda', confidence_threshold=0.5):
    """Predict pigment segmentation for a single image"""

    # Preprocess image
    image_tensor, original_image = preprocess_image(image_path)
    image_tensor = image_tensor.to(device)

    # Inference
    with torch.no_grad():
        outputs = model(image_tensor)

        # Get probability map
        probabilities = torch.softmax(outputs, dim=1)
        pigment_prob = probabilities[0, 1, :, :].cpu().numpy()  # Class 1 (pigment)

        # Get prediction mask
        prediction = (pigment_prob > confidence_threshold).astype(np.uint8)

    return original_image, pigment_prob, prediction

def visualize_results(original_image, probability_map, prediction_mask,
                     image_path, save_path=None, confidence_threshold=0.5):
    """Visualize the segmentation results"""

    # Resize probability map and prediction to original image size
    orig_h, orig_w = original_image.shape[:2]
    probability_map_resized = cv2.resize(probability_map, (orig_w, orig_h))
    prediction_mask_resized = cv2.resize(prediction_mask, (orig_w, orig_h), interpolation=cv2.INTER_NEAREST)

    # Create visualization
    fig, axes = plt.subplots(2, 3, figsize=(15, 10))

    # Original image
    axes[0, 0].imshow(original_image)
    axes[0, 0].set_title('Original Image', fontsize=12, fontweight='bold')
    axes[0, 0].axis('off')

    # Probability heatmap
    im1 = axes[0, 1].imshow(probability_map_resized, cmap='jet', vmin=0, vmax=1)
    axes[0, 1].set_title('Pigment Probability Map', fontsize=12, fontweight='bold')
    axes[0, 1].axis('off')
    plt.colorbar(im1, ax=axes[0, 1], fraction=0.046, pad=0.04)

    # Binary prediction mask
    axes[0, 2].imshow(prediction_mask_resized, cmap='gray')
    axes[0, 2].set_title(f'Binary Mask (threshold={confidence_threshold})', fontsize=12, fontweight='bold')
    axes[0, 2].axis('off')

    # Overlay on original
    overlay = original_image.copy()
    overlay[prediction_mask_resized == 1] = [255, 0, 0]  # Red for pigment
    blended = cv2.addWeighted(original_image, 0.7, overlay, 0.3, 0)
    axes[1, 0].imshow(blended)
    axes[1, 0].set_title('Prediction Overlay (Red)', fontsize=12, fontweight='bold')
    axes[1, 0].axis('off')

    # Contour visualization
    contour_img = original_image.copy()
    contours, _ = cv2.findContours(prediction_mask_resized, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
    cv2.drawContours(contour_img, contours, -1, (255, 0, 0), 2)
    axes[1, 1].imshow(contour_img)
    axes[1, 1].set_title('Contour Detection', fontsize=12, fontweight='bold')
    axes[1, 1].axis('off')

    # Statistics
    total_pixels = orig_h * orig_w
    pigment_pixels = np.sum(prediction_mask_resized == 1)
    pigment_percentage = (pigment_pixels / total_pixels) * 100
    max_prob = np.max(probability_map_resized)
    mean_prob = np.mean(probability_map_resized)

    stats_text = f"""
    Image: {os.path.basename(image_path)}

    Statistics:
    • Total pixels: {total_pixels:,}
    • Pigment pixels: {pigment_pixels:,}
    • Pigment coverage: {pigment_percentage:.2f}%
    • Max probability: {max_prob:.3f}
    • Mean probability: {mean_prob:.3f}
    • Threshold: {confidence_threshold}
    • Detected regions: {len(contours)}
    """

    axes[1, 2].text(0.05, 0.95, stats_text, transform=axes[1, 2].transAxes,
                    fontsize=10, verticalalignment='top', fontfamily='monospace',
                    bbox=dict(boxstyle='round', facecolor='wheat', alpha=0.8))
    axes[1, 2].axis('off')

    plt.suptitle('Retinal Pigment Segmentation Results', fontsize=16, fontweight='bold')
    plt.tight_layout()

    if save_path:
        plt.savefig(save_path, dpi=300, bbox_inches='tight')
        print(f"Results saved to: {save_path}")

    plt.show()

    return {
        'pigment_percentage': pigment_percentage,
        'max_probability': max_prob,
        'mean_probability': mean_prob,
        'detected_regions': len(contours),
        'total_pixels': total_pixels,
        'pigment_pixels': pigment_pixels
    }

def test_single_image(image_path, model_path='/content/best_ultra_stable_rps_net.pth',
                     confidence_threshold=0.5, save_results=True):
    """Complete pipeline to test a single image"""

    # Check if image exists
    if not os.path.exists(image_path):
        print(f"❌ Image not found: {image_path}")
        return None

    # Set device
    device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
    print(f"Using device: {device}")

    # Load model
    model = load_trained_model(model_path, device)
    if model is None:
        return None

    try:
        # Predict
        print(f"\n🔍 Processing image: {os.path.basename(image_path)}")
        original_image, probability_map, prediction_mask = predict_single_image(
            model, image_path, device, confidence_threshold
        )

        # Visualize
        save_path = None
        if save_results:
            base_name = os.path.splitext(os.path.basename(image_path))[0]
            save_path = f"segmentation_result_{base_name}.png"

        stats = visualize_results(
            original_image, probability_map, prediction_mask,
            image_path, save_path, confidence_threshold
        )

        print(f"\n📊 Analysis Results:")
        print(f"  - Pigment coverage: {stats['pigment_percentage']:.2f}%")
        print(f"  - Detected regions: {stats['detected_regions']}")
        print(f"  - Max confidence: {stats['max_probability']:.3f}")
        print(f"  - Mean confidence: {stats['mean_probability']:.3f}")

        return stats

    except Exception as e:
        print(f"❌ Error processing image: {e}")
        import traceback
        traceback.print_exc()
        return None

def batch_test_images(image_dir, model_path='/content/best_ultra_stable_rps_net.pth',
                     confidence_threshold=0.5, image_extensions=['.jpg', '.jpeg', '.png']):
    """Test multiple images in a directory"""

    if not os.path.exists(image_dir):
        print(f"❌ Directory not found: {image_dir}")
        return

    # Find all images
    image_files = []
    for ext in image_extensions:
        image_files.extend([f for f in os.listdir(image_dir) if f.lower().endswith(ext)])

    if not image_files:
        print(f"❌ No images found in {image_dir}")
        return

    print(f"Found {len(image_files)} images")

    # Load model once
    device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
    model = load_trained_model(model_path, device)
    if model is None:
        return

    # Process each image
    results = []
    for img_file in image_files:
        img_path = os.path.join(image_dir, img_file)
        print(f"\n📷 Processing: {img_file}")

        try:
            original_image, probability_map, prediction_mask = predict_single_image(
                model, img_path, device, confidence_threshold
            )

            # Calculate stats
            orig_h, orig_w = original_image.shape[:2]
            prediction_mask_resized = cv2.resize(prediction_mask, (orig_w, orig_h), interpolation=cv2.INTER_NEAREST)

            pigment_pixels = np.sum(prediction_mask_resized == 1)
            pigment_percentage = (pigment_pixels / (orig_h * orig_w)) * 100
            max_prob = np.max(probability_map)

            results.append({
                'filename': img_file,
                'pigment_percentage': pigment_percentage,
                'max_probability': max_prob,
                'pigment_pixels': pigment_pixels
            })

            print(f"  ✅ Pigment coverage: {pigment_percentage:.2f}%, Max prob: {max_prob:.3f}")

        except Exception as e:
            print(f"  ❌ Error: {e}")
            results.append({
                'filename': img_file,
                'error': str(e)
            })

    # Summary
    successful_results = [r for r in results if 'error' not in r]
    if successful_results:
        avg_coverage = np.mean([r['pigment_percentage'] for r in successful_results])
        print(f"\n📊 Batch Summary:")
        print(f"  - Processed: {len(successful_results)}/{len(image_files)} images")
        print(f"  - Average pigment coverage: {avg_coverage:.2f}%")

    return results

# Example usage
if __name__ == "__main__":
    # Test single image
    # Replace with your actual image path
    image_path = "/content/Fundus_of_patient_with_retinitis_pigmentosa,_mid_stage.jpg"

    # Uncomment to test single image:
    stats = test_single_image(image_path, confidence_threshold=0.5)

    # Uncomment to test batch of images:
    # results = batch_test_images("/path/to/your/image/directory")

    print("🎯 Single Image Testing Code Ready!")
    print("\nTo use:")
    print("1. test_single_image('path/to/image.jpg') - Test one image")
```
